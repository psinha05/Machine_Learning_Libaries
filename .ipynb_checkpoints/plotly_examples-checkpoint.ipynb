#  Matplot Library

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np 
```

## Pie Chart

```python
areas = ['HR', 'Marketing', 'Sales', 'IT', 'Customer Support']
budget = [1.2, 2.2, 3.1, 7.2, 3.3]   

plt.pie(budget, labels=areas, autopct= "%0.1f%%")
plt.show()
```

```python
# Scatter plot

x=[7, 11, 15, 22, 19, 55, 33]
y=[5, 15, 39, 7, 1, 30, 77 ]

plt.scatter(x, y)
```

```python
# Scatter plot
    
X=[7, 11, 15, 22, 19, 55, 33]
Y=[15, 39, 7, 1, 30, 77, 98 ]

plt.scatter(X, Y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age Vs Salary Graph')
```

```python
# 2-D plot 
product_id = [101, 102, 103, 104, 105, 106, 107, 108]
online_price = [200, 295, 320, 497, 519, 285, 100, 770]
offline_price = [150, 235, 220, 397, 419, 225, 90, 750]
chor_bazaar = [100, 195, 220, 297, 190, 570, 80, 700]
plt.plot(product_id, online_price)
```

```python
plt.plot(product_id, online_price, color='green', marker ='+')
plt.xlabel('Product')
plt.ylabel('Price')
plt.title('Product Price Comparison')
plt.show()
```

```python
plt.plot(product_id, online_price, color='green', marker ='D', markersize= 10, label ='Online Prices')
plt.plot(product_id, offline_price, color='blue', marker = '+', markersize = 12, label='Offline Prices')
plt.plot(product_id, chor_bazaar, color='yellow', marker = '*', markersize =14, label='Chor Bazaar')
plt.xlabel('Product')
plt.ylabel('Price')
plt.legend()
plt.title('Product Price Comparison')
plt.show()
```

```python
# bar plot vertically plotted 

Year = [2012, 2013, 2014, 2015, 2016, 2017, 2018]
Price = [45000, 50000, 60000, 52000, 58000, 65000, 82000]
plt.bar(Year, Price, color='grey')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Bike Price Trends in Years')
plt.show()
```

```python
# bar plot horizontally plotted 

Year = [2012, 2013, 2014, 2015, 2016, 2017, 2018]
Price = [45000, 50000, 60000, 52000, 58000, 65000, 82000]
plt.barh(Year, Price, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Bike Price trends over the years')
plt.show()
```
