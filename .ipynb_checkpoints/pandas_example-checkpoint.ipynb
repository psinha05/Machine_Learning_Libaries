# Merging an data frame

```python
# employee dataframe
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
# Merge based on the employee_id in boh the data frames 
pd.merge(employees, salaries, on='employee_id')
```

```python
# based on the inner(common keys in both the data frames) 
pd.merge(employees, salaries, on='employee_id', how='inner')
```

```python
# merging based on outer join, i.e, employee_id : to contains all the employee_id records 
pd.merge(employees, salaries, on='employee_id', how='outer')
```

```python
# left join, merge based on employee(employee_id)
pd.merge(employees, salaries, on='employee_id', how='left')
```

```python
# right join, merge based on salaries 
pd.merge(employees, salaries, on='employee_id', how='right')
```

# Pivot Tables 
# Summarize the substantial tables, it may include mean, median, sum, or other statistical terms 

```python
# importing pandas
import pandas as pd

# creating dataframe
df = pd.DataFrame({'Product': ['Carrots', 'Broccoli', 'Banana', 'Banana',
                               'Beans', 'Orange', 'Broccoli', 'Banana'],
                   'Category': ['Vegetable', 'Vegetable', 'Fruit', 'Fruit',
                                'Vegetable', 'Fruit', 'Vegetable', 'Fruit'],
                   'Quantity': [8, 5, 3, 4, 5, 9, 11, 8],
                   'Amount': [270, 239, 617, 384, 626, 610, 62, 90]})
```

```python
df
```

```python
# Get the total sales of each products 
df.pivot_table(index= ['Product'],
               values =['Amount'],
               aggfunc='sum')
```

```python
# Get the total sales of each category
df.pivot_table(index=['Category'],
               values =['Amount'],
               aggfunc='sum')
```

```python
#  Total Sales by Category and Product Both

df.pivot_table(index=['Product','Category'],
               values =['Amount'],
               aggfunc='sum')
```

```python
#  Get the Mean, Median, Minimum Sale by Category

df.pivot_table(index=['Category'],
               values =['Amount'],
               aggfunc={'median', 'mean', 'min'})
```

```python
# Get the Mean, Median, Minimum Sale by Product
df.pivot_table(index=['Product'],
               values =['Amount'],
               aggfunc={'median', 'mean', 'min'})
```

```python
# to extract the no. of episode from each rows.

def extract_episodes(txt):
    check = False
    data = ""
    
    for i in txt:
        if i == ")":
            check = False
            break
        if check == True:
            data += i
        if i == "(":
            check = True
    return data
```

```python
#  check = True starts collecting characters.
#  check = False stops collecting.
# You use this to pull out only what's inside the parentheses.

| Character (`i`) | What happens?                         | `check` | `data`        |
| --------------- | ------------------------------------- | ------- | ------------- |
| B, r, e, a...   | Not `(`, so nothing collected.        | False   | ""            |
| `(`             | Sets `check = True`                   | True    | ""            |
| `6`             | Since `check == True`, adds to `data` | True    | "6"           |
| `2`             | `check == True`, adds to `data`       | True    | "62"          |
| space, e, p...  | keep adding                           | True    | "62 episodes" |
| `)`             | Sets `check = False`, breaks loop     | False   | done          |

```

```python
df_anime["Title"].apply(extract_episodes)
```

```python
# to create a new column and store the episodes in "Episode" columns
df_anime["Episodes"]= df_anime["Title"].apply(extract_episodes)
```

```python
import pandas as pd
from dateutil import parser
from dateutil.relativedelta import relativedelta
import re
```

```python
def extraction_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+20):
                data += txt[j]

            return data
```

```python
df_anime['Total Duration'] = df_anime['Title'].apply(extraction_time)
```

```python
df_anime.head()
```

```python
# To convert the Total Duration into Months(int)
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None

df_anime['Months'] = df_anime['Total Duration'].apply(calculate_total_months)
```
